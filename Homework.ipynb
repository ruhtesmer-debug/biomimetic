## How Comments Work

In Python, a **comment** is text that Python ignores — it's there for humans to read, not for the computer to execute.

Comments start with a `#`. Everything after the `#` on that line is ignored:

```python
x = 5  # this is a comment — Python ignores everything after the #
# this entire line is a comment
```

In this notebook, code cells often include comments like:

